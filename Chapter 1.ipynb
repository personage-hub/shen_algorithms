1.1.1. Даны две целые переменные a, b. Составьте фрагмент программы, после исполнения которого значения переменных поменялись бы местами.

```python
a = int(input("Введите переменную a "))
b = int(input("Введите переменную b "))
print(f"До выполнения\nПеременная a = {a}\nПеременная b = {b}")
# введем дополнительную переменную t для хранения данных и просвоим значение одной из переменных
t = a
# выполним операции присваивания
a = b
b = t
# выводим результат
print(f"После выполнения\nПеременная a = {a}\nПеременная b = {b}")
```

1.1.2. Решить предыдущую задачу без использования массивов

```python
a = int(input("Введите переменную a "))
b = int(input("Введите переменную b "))
print(f"До выполнения\nПеременная a = {a}\nПеременная b = {b}")
# Суммируем значения двух переменных и почердно вычтем 
a = a + b # теперь в a записано значение суммы двух перменных {a = a0 + b0; b = b0}
b = a - b # вычитая из суммы двух перменных первоначальное значение b, получаем новое значение b {a = a0 + b0; b = a0}
a = a - b # вычитая из суммы двух перменных новое значение b, получаем новое значеие a {a = b0; b = a0}
# выводим результат
print(f"После выполнения\nПеременная a = {a}\nПеременная b = {b}")
```

1.1.3. Даное целое число а и натуральное число n. Необходимо составить программу, вычисляющее значение переменной b, которое становится равным a^n.

```python
a = int(input("Введите переменную a "))
n = int(input("Введите положительное значение переменной n "))
b = 1
while n < 0: # Проверяем условие ввода
    print("Вы ввели отрицательное значение")
    n = int(input("Введите положительное значение переменной n "))
# введем промежуточную переменную k, значение которой меняется от 0 до n
k = 0
i = 0 # Служебная переменная для подсчёта количетсва циклов
while k < n:
    k += 1
    b *= a
    i += 1
print(f"Значение переменной b (a ^ b) = {b}")
print(f"Количество циклов = {i}")
```

Другое решение этой же задачи:

```python
a = int(input("Введите переменную a "))
n = int(input("Введите положительное значение переменной n "))
b = 1
while n < 0: # Проверяем условие ввода
    print("Вы ввели отрицательное значение")
    n = int(input("Введите положительное значение переменной n "))
# введем промежуточную переменную k, значение которой меняется от n до 0
k = n
i = 0 # Служебная переменная для подсчёта количетсва циклов
while k != 0:
    k -= 1
    b *= a
    i +=1
print (f"Значение переменной b (a ^ b) = {b}")
print (f"Количество циклов = {i}")
```

1.1.4. Решить предыдущую задачу с ограничением количества циклов порядко log(n), т.е. было бы не более C * log (n) для некоторой константы C.

```python
# Решение основано на втором методе решения (с уменьшением служебной переменной)
a = int(input("Введите переменную a "))
n = int(input("Введите положительное значение переменной n "))
b = 1
while n < 0: # Проверяем условие ввода
    print("Вы ввели отрицательное значение")
    n = int(input("Введите положительное значение переменной n "))
# введем промежуточную переменную k, значение которой меняется от n до 0
k = n
# введем служебную переменную c для поэтамного решения уравнения {a^n = b*(c^k)}
c = a
i = 0 # Служебная переменная для подсчёта количетсва итераций
while k != 0:
    i += 1
    if k % 2 == 0: # Если k чётное
        k /= 2 # делим k попоплам
        c *= c
        print(f"\tПромежуточное значение c = {c}")
    elif k % 2 != 0:
        k -= 1 # уменьшаем k, чтобы получить четное число
        b *= c # умножаем b на c
        print(f"\tПромежуточное значение b * c =  {b}")
print (f"\n\n Итоговое значение переменной b (a ^ b) = {b}")
print (f"Количество итераций = {i}")
```

Каждый второй раз (не реже) будет выполняется код для четного значения k, таким образом за две итерации k уменьшается по крайней мере вдвое. Таким образом уменьшаем количество итераций. Например, при возведение в степень 50 требуется 8 итераций.
